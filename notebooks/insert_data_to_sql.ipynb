```python
import pandas as pd
import mysql.connector

file_path = r"C:\Users\ujjwa\Downloads\complaint-intelligence-platform\data\processed\complaints_mysql_safe.csv"

# Load CSV
df = pd.read_csv(file_path, low_memory=False)

# Reorder columns to match SQL insert order
df = df[
    [
        "complaint_id",
        "date_received",
        "product",
        "sub_product",
        "issue",
        "sub_issue",
        "complaint_text",
        "company",
        "state",
        "submitted_via",
        "timely_response",
        "consumer_disputed",
    ]
]

# Convert NaN to None
df = df.where(pd.notnull(df), None)

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="localhost",
    database="complaint_analytics"
)

cursor = conn.cursor()

insert_query = """
INSERT INTO complaints
(complaint_id, date_received, product, sub_product, issue, sub_issue,
 complaint_text, company, state, submitted_via, timely_response, consumer_disputed)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

batch_size = 5000
rows = [tuple(row) for row in df.itertuples(index=False, name=None)]

for i in range(0, len(rows), batch_size):
    batch = rows[i:i + batch_size]
    cursor.executemany(insert_query, batch)
    conn.commit()
    print(f"Inserted rows: {i + len(batch)}")

cursor.close()
conn.close()

print("Import completed successfully.")
```
